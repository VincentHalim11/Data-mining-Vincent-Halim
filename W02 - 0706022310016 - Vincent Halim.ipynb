```python
#vincent Halim - 0706022310016
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url,sep=',')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
null_sum = df.isnull().sum()
null_perc = (null_sum/len(df))*100

null_df = pd.DataFrame({'Null Count' : null_sum, 'Null Percentages' : null_perc})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentages'] = null_df['Null Percentages'].map('{:.2f}%'.format)
null_df
```

```python
#dari dataset ini dapat disimpulkan jika ada 3552 data tetapi juga ditemukan banyak data yang kosong
#terutama di bagian building_orientation hingga 1647 data selain itu masih ada building_age dan year_built yang sama 1445 data
#tetapi ada juga yang hanya kosong sedikit contohnya di porperty type yang hanya memiliki 1 data kosong
```

MISSING DATA

```python
# data kosong bagian ini aku memilih untuk mengisi data kosong dengan mediannya. karena aku merasa cukup banyak jika harus didrop
col = ["bedrooms","bathrooms","land_size_m2","building_size_m2","floors"]
for cols in col:
    df[cols].fillna(df[cols].median(), inplace= True)
```

```python
#aku merasa ini juga memilih untuk mengisi data ini akan tetapi aku memilih untuk mengisinya dengan "unknown" karena isi data bukan
# int  aku melakukan ini untuk mencegah kehilangan banyak data
kol = ["certificate","building_age","year_built","property_condition","building_orientation","furnishing"]
df[kol] = df[kol].fillna("Unknown")
```

```python
#aku memilih untuk menghapus data ini karena data ini kecil hanya 1 hingga 4 data saja dan karena data awal ada 3552 data aku rasa
#kehilangan 1-4 data tidak memberikan pengruh yang besar
df = df.dropna(subset=["ads_id","property_type"])
```

```python
null_sum = df.isnull().sum()
null_perc = (null_sum/len(df))*100

null_df = pd.DataFrame({'Null Count' : null_sum, 'Null Percentages' : null_perc})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentages'] = null_df['Null Percentages'].map('{:.2f}%'.format)
null_df
```

```python
# binning
bins = [0, 2, 4, 6, 8, 10, df["bedrooms"].max()]
labels = ["0-2", "3-4", "5-6", "7-8", "9-10", "10+"]
df["bedrooms_bin"] = pd.cut(df["bedrooms"], bins=bins, labels=labels, include_lowest=True)


plt.figure(figsize=(7,5))
sn.countplot(x="bedrooms_bin", data=df, palette="Blues")
plt.title("Binning Barchart Bedrooms")
plt.xlabel("Jumlah Bedrooms (bin)")
plt.ylabel("Jumlah Properti")
plt.show()
```

```python
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
city_counts = df["city"].value_counts()
print(city_counts)

plt.figure(figsize=(10,5))
sn.countplot(data=df, x="city", order=df["city"].value_counts().index)

plt.title("Jumlah Properti per Kota")
plt.xticks(rotation=45)
plt.show()
#disini saya menggunakan barchart  untuk mencari tau banyak rumah yang tersebar di kota-kota dimana ditemukan 881 rumah terletak di bogor
#dan kota dengan rumah terbanyak ada di tanggerang dengan sebesar 848 dan disusul dengan kota bekasi dengan jumlah yang cukup menurun jika
#dibandingkan dengan kota - kota sebelumnya yaitu di 586 rumah
```

```python
condition_counts = df["property_condition"].value_counts()
print(condition_counts)


df["property_condition"].value_counts().plot(
    kind="barh",
    figsize=(8,6),
    color="skyblue"
)

plt.title("Distribusi Kondisi Properti")
plt.xlabel("Jumlah")
plt.ylabel("Kondisi")
plt.show()
#disini saya menganalisis kondisi properti dimana ditemukan kondisi paling banyak adalah rumah dengan kondisi bagus dengan sebanyak 1437 rumah
# disusul dengan rumah dengan kondisi baru sebanyak 1328 rumah. sedangkan yang paling sedikit adalah rumah dengan kondisi semi furnished
# hanya dengan 4 rumah saja.
```

```python
furnishing_counts = df["furnishing"].value_counts()
print(furnishing_counts)

df["furnishing"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6),
    startangle=90
)

plt.ylabel("")
plt.title("Distribusi Kondisi Properti")
plt.show()

# disini adakah distribusi kondisi properti dengan menggunakan pie chart. dimana didapatkan banyak rumah dengan kondisi unfurnished dengan
# persentase sebanyak 58.2% disusul dengan rumah dengan kondisi semi - furnished dengan 23.5% dan yang pling sedikit adalah rumah dalam
# kondisi baru dimana hanya 0.8% saja
```

```python
cols = ["maid_bedrooms", "maid_bathrooms"]

for col in cols:
    plt.figure(figsize=(5,4))
    plt.boxplot(df[col].dropna())
    plt.title(f"Boxplot {col}")
    plt.ylabel(col)
    plt.show()

# dari boxplot ini didapatkan jika rumah - rumah kebanyakan tidak memiliki kamar mandi dan kamar pembantu tetapi muncul outlier yang
# signifikan menunjukan rumah - rumah yang memiliki kamar dan kamar mandi pembantu adalah rumah mewah
```
